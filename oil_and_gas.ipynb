## Yearly Gas production

```python
yearly_gas_df = gas_df.groupby(gas_df["Month"].dt.year).sum(numeric_only=True)
```

### Check your code

```python
index_year = yearly_gas_df.index[0]
yearly_gas_shape = yearly_gas_df.shape
us_total = yearly_gas_df.iloc[0,0]
```

```python
from nbresult import ChallengeResult

result = ChallengeResult('full_gas',
    index_year=index_year,
    yearly_gas_shape=yearly_gas_shape,
    us_total=us_total
)
result.write()

print(result.check())
```

```python
plot = yearly_gas_df.filter(items=['U.S.']).plot(kind="bar")
plot.set_xlabel("Year");
```

```python
first_year = yearly_gas_df.index.min()
last_year = yearly_gas_df.index.max()
```

```python
year_counts = gas_df.groupby(gas_df["Month"].dt.year)["Month"].count()
year_counts
```

```python
valid_years = year_counts[year_counts == 12].index
valid_years
```

```python
import numpy as np

condition = np.logical_and(
    yearly_gas_df.index >= valid_years.min(),
    yearly_gas_df.index <= valid_years.max()
)
```

```python
filtered_yearly_gas_df = yearly_gas_df[condition]
filtered_yearly_gas_df
```

```python
import matplotlib.pyplot as plt

filtered_yearly_gas_df["U.S."].plot(kind="bar")
plt.title("U.S. Yearly Gas Production (Complete Years Only)")
plt.xlabel("Year")
plt.ylabel("Total Production")
plt.show()
```

### Check your code

## Comparing with Crude Oil Production

```python
import pandas as pd

file = "data/U.S._crude_oil_production.csv"
oil_df = pd.read_csv(file, decimal=",", parse_dates=['Month'])
oil_df.head(3)
```

```python
oil_df["Month"].dtype
```

---

## Yearly Oil production

```python
oil_df['Year'] = oil_df['Month'].dt.year
```

```python
oil_df.columns = oil_df.columns.str.strip()
```

```python
yearly_oil_df = oil_df.groupby('Year').sum(numeric_only=True)
yearly_oil_df
```

```python
print(yearly_oil_df.columns)
```

```python
import matplotlib.pyplot as plt

yearly_oil_df.filter(items=['U.S. Crude Oil']).plot(kind='bar');
```

```python
first_year = yearly_oil_df.index.min()
last_year = yearly_oil_df.index.max()
```

```python
year_counts = oil_df.groupby(oil_df["Month"].dt.year)["Month"].count()
year_counts
```

```python
valid_years = year_counts[year_counts == 12].index
valid_years
```

```python
import numpy as np

condition = np.logical_and(
    yearly_oil_df.index >= valid_years.min(),
    yearly_oil_df.index <= valid_years.max()
)
```

```python
filtered_yearly_oil_df = yearly_oil_df[condition]
filtered_yearly_oil_df
```

### Check your code

```python
from nbresult import ChallengeResult

result = ChallengeResult('oil',
    filtered_oil_shape=filtered_yearly_oil_df.shape,
    filtered_oil_index_year=filtered_yearly_oil_df.index[0],
    us_total=filtered_yearly_oil_df.iloc[0,0]
)
result.write()

print(result.check())
```

## Merging Oil and Gas Production

```python
import pandas as pd

#YalnÄ±zca toplam ABD Ã¼retimi sÃ¼tunlarÄ±nÄ± seÃ§elim
total_oil = filtered_yearly_oil_df[['U.S. Crude Oil']].copy()
total_gas = filtered_yearly_gas_df[['U.S.']].copy()  
#SÃ¼tun adlarÄ±nÄ± yeniden adlandÄ±ralÄ±m
total_oil.rename(columns={'U.S. Crude Oil': 'Crude Oil'}, inplace=True)
total_gas.rename(columns={'U.S.': 'Gas'}, inplace=True)

#Ä°kisini birleÅŸtirelim (axis=1 => yan yana)
merged_df = pd.concat([total_oil, total_gas], axis=1)

#Sonucu kontrol edelim
print(merged_df.head())
```

```python
plot = merged_df.plot(kind="bar")
plot.set_xlabel("Year")
plot.legend(['Gas (Millions of Cubic feet)', 'Crude Oil (Thousands of barrels)']);
```
